Desde la dirección se están planteando cambiar el sistema de asiganción de las solicitudes que llegan. En concreto, necesitan saber que categoría es la que recibe más tickets para poder asignar más empleados a esa categoría. Por esto nos piden que averiguemos qué categoría es la que tiene más tickets. Además nos piden que devolvamos esta consulta en un DataFrame.

📌 NOTA Ordena los resultados de mayor a menor. Esto todavía no lo hemos visto, pero un DataFrame lo podemos ordenar con .sort_values(). Aquí   tenéis una explicación de cómo hacerlo.

```python
df_tickets = pd.read_csv("datos/tickets_years.csv")
```

```python
df_tickets["categoría"].unique()
```

```python
df_tickets_categoria = df_tickets.groupby("categoría")["id_ticket"].count().reset_index()
```

```python
df_tickets_categoria = df_tickets_categoria.sort_values(by = "id_ticket", ascending = False)
```

```python
df_tickets_categoria
```

2. Desde la dirección quieren seguir evaluando la situación planteada en el ejercicio anterior. Pero ahora quieres saber cual es la media de la satisfacción por categoría. En concreto quieren saber si independientemente de la categoría los clientes salen satisfechos, para que en caso de que no sea así poder identificar un posible problema subyacente. Por lo tanto, nos piden que calculemos la media de satisfacción por cada una de las categorías que tenemos en el dataframe.

```python
df_satisfaccion_categoria = df_tickets.groupby("categoría")["satisfacción"].mean().reset_index()
df_satisfaccion_categoria
```

La media de satisfacción es aproximadamente 4 para todas las categorías.

3. Quieren mejorar la consulta anterior y ahora nos piden que además de la media les devolvamos la desviación estándar de la satisfacción.

```python
df_satisfaccion_stats = df_tickets.groupby("categoría")["satisfacción"].agg(media_de_satisfaccion = ("mean"), desviacion_estandar_satisfaccion = ("std")).reset_index()
df_satisfaccion_stats
```

4. En nuestros datos, además hemos visto que no solo tenemos categorías, y que las diferencias entre ellas no son especialmente significativas. Revisando los datos de nuevo, desde dirección saben que hay dos tipos de tickets (Problema y Solicitud), basándonos en esto, ¿hay diferencias en la satisfacción basándonos en la categoría y el tipo de tickets?.

```python
df_sat_categoria_tipo = df_tickets.groupby(["tipo", "categoría"])["satisfacción"].mean().reset_index()
df_sat_categoria_tipo
```

No vemos grandes diferencias en la media de satisfacción por categoría y tipo.
