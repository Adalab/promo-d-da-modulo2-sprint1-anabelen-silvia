```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("datos/2016_edit.csv")
```

```python
df_soft = df[(df["categor√≠a"]  == "Software")]
df_hard = df[(df["categor√≠a"]  == "Hardware")]
```

```python
df_tipo_soft = df_soft["tipo"].value_counts().reset_index()
df_tipo_hard = df_hard["tipo"].value_counts().reset_index()
```

```python
df_sat_soft = df_soft["satisfacci√≥n"].value_counts().reset_index()
df_sat_hard = df_hard["satisfacci√≥n"].value_counts().reset_index()
```

```python
# a√±adimos el par√°metro figsize para establecer el tama√±o de la gr√°fica
fig,ax = plt.subplots(1,2, figsize =( 15,5))

ax[0].bar(df_tipo_soft["index"], df_tipo_soft["tipo"], edgecolor = "black")
ax[0].set_xlabel("Tipos de incidencias")
ax[0].set_ylabel("N√∫mero de tickets")
ax[0].set_title("Software")


ax[1].bar(df_tipo_hard["index"], df_tipo_hard["tipo"], edgecolor = "black", color = "lime")
ax[1].set_xlabel("Tipos de incidencias")
ax[1].set_ylabel("N√∫mero de tickets")
ax[1].set_title("Hardware")
plt.savefig("hardware_software.jpg")
plt.show()
```

```python
# a√±adimos el par√°metro figsize para establecer el tama√±o de la gr√°fica
fig,ax = plt.subplots(2,1, figsize =(6,10))

ax[0].bar(df_sat_soft["index"], df_sat_soft["satisfacci√≥n"], edgecolor = "black", color = "maroon")
ax[0].set_xlabel("Nivel de satisfacci√≥n")
ax[0].set_ylabel("N√∫mero de tickets")
ax[0].set_title("Software")


ax[1].bar(df_sat_hard["index"], df_sat_hard["satisfacci√≥n"], edgecolor = "black", color = "lime")
ax[1].set_xlabel("Nivel de satisfacci√≥n")
ax[1].set_ylabel("N√∫mero de tickets")
ax[1].set_title("Hardware")
plt.savefig("satisfaccion_tipos.jpg")
plt.show()
```

Cread un subplot con dos gr√°ficas, d√≥nde en la primera tengamos el n√∫mero de empleados por departamento y en la otra por tipo de puesto. Deb√©is:

    - Poner leyenda a la gr√°fica
    - Cada l√≠nea debe ir en un color
    - Ponerle nombre a los ejes y a la gr√°fica.

```python
df_empleo = pd.read_csv("datos/empleo.csv")
```

```python
df_empleo.head(10)
```

```python
df_departamento = df_empleo["departamento"].value_counts().reset_index()
df_departamento
```

```python
df_puesto = df_empleo["tipo_puesto"].value_counts().reset_index()
df_puesto
```

```python
# a√±adimos el par√°metro figsize para establecer el tama√±o de la gr√°fica
fig,ax = plt.subplots(2,1, figsize =(15,10))

ax[0].plot(df_departamento["index"], df_departamento["departamento"], marker = "v")
ax[0].set_xlabel("Departamentos")
ax[0].tick_params(rotation = 45)
ax[0].set_ylabel("N√∫mero de empleados")
ax[0].set_title("Departamento")
ax[0].legend( handles=["N√∫mero de empleados"], labels = "empleados")


ax[1].plot(df_puesto["index"], df_puesto["tipo_puesto"], color = "lime", marker = "D")
ax[1].set_xlabel("Tipos de puestos")
ax[1].set_ylabel("N√∫mero de empleados")
ax[1].set_title("Tipo de puesto")
fig.tight_layout()
plt.savefig("grafica_empleados.jpg")
plt.show()
```
