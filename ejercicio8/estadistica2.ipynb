```python
df = pd.read_csv("avocado.csv", index_col=0)
```

```python
df.head()
```

```python
df.describe().T
```

```python
print(f"la mediana es :" , df["AveragePrice"].median())
print(f"la mediana es :" , df["Total Volume"].median())
print(f"la mediana es :" , df["year"].median())
```

```python
print(f"la moda es :" , df["AveragePrice"].mode())
print(f"la moda es :" , df["Total Volume"].mode())
print(f"la moda es :" , df["year"].mode())
```

```python
df.describe().T
```

```python
df["precio_medio"] = pd.cut(x = df["AveragePrice"], bins = [0.4, 1.10, 1.38, 1.70, 3.26], labels = ["Bajo", "Medio", "Alto", "Muy_alto"])
```

```python
df["volumen_compra"] = pd.cut(x = df["Total Volume"], bins = [84.50, 10840, 108000, 500000, 62506000], labels = ["Bajo", "Medio", "Alto", "Muy_alto"])
```

```python
df.head()
```

```python
df["precio_medio"].value_counts()
```

```python
df["volumen_compra"].value_counts()
```

```python
df1 = df.groupby(["precio_medio", "volumen_compra"])["Date"].count().reset_index()
```

```python
df1
```

Aqui tenemos la tabla de frecuencias comparando los precios y el volumen de ventas; a mayor volumen, menor precio.

```python
df2 = df.groupby(["year"]).count().reset_index().sort_values(by="Total Volume", ascending = False)
```
