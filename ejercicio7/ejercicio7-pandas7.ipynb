A la hora de escribir la funci칩n debemos tener en cuenta que puede que no haya la informaci칩n de mes en la columna de la fecha. En ese caso devolved un nulo (NaN).
游늷 Nota La fecha tendr치 que ir formato de string como "Enero", "Febrero", etc.

```python
import pandas as pd
import numpy as np
import re
import datetime as dt
```

```python
df_tickets = pd.read_csv("datos/tickets_years.csv", index_col = 0)
df_tickets.head()
```

```python
def sacar_mes(mes):
    try:
        return mes.strftime("%B")
    except: 
        return np.nan
```

```python
df_tickets["fecha"]= df_tickets["fecha"].apply(pd.to_datetime)
```

```python
df_tickets["mes"]= df_tickets["fecha"].apply(sacar_mes)
```

```python
df_tickets.head()
```

```python
diccionario_mes = {"January" : "Enero", "February": "Febrero", "March": "Marzo", "April": "Abril", "May": "Mayo", "June": "Junio", "July": "Julio", "August": "Agosto", "September": "Septiembre", "October": "Octubre", "November": "Noviembre", "December": "Diciembre"}
```

```python
diccionario_mes["April"]
```

```python
def traducir_fecha (columna):
    try:
        return diccionario_mes[columna]
    except: 
        return np.nan    
```

```python
df_tickets["mes"]= df_tickets["mes"].apply(traducir_fecha)
df_tickets.head()
```

    2. En la tabla de empleados, tenemos los nombres y los apellidos de los empleados, pero en columnas separadas. En los ejercicios de pair de la lecci칩n del groupby vimos que si quisieramos agrupar por los nombres, esto se nos puede complicar. Ahora nos piden que creemos una columna donde tengamos el nombre y el apellido juntos.
游늷 NOTA La columna nueva debe tener el nombre y apellido con la primera letra en may칰sculas, por ejemplo "Jes칰s Villa".
游늷 NOTA Una vez hallamos creado la columna nueva, eliminad las columnas de nombre y apellido originales.


```python
df_empleados = pd.read_csv("datos/empleo.csv", index_col = 0)
df_empleados.head()
```

```python
def juntar_nombre (nombre, apellido):
    nombre= nombre.title()
    apellido= apellido.title()
    nombre_completo = nombre + " " + apellido
    return nombre_completo
```

```python
df_empleados["nombre_completo"] = df_empleados.apply(lambda dataframe: juntar_nombre(dataframe["nombre"], dataframe["apellido"]), axis = 1)
```

```python
df_empleados.head()
```

```python
df_empleados.drop(["apellido", "nombre"], axis = 1, inplace= True)
```

En el pair programming desde a direcci칩n nos ped칤an que agruparamos los datos en base a la columna de "categor칤a", puede que en aquel momento nos dieramos cuenta que hab칤a var칤as categor칤as que eran igual (en concreto "Acceso Login") pero que en algunos casos ten칤a unos s칤mbolos raros (&, %, -). Es el momento de limpiar esta columna. El objetivo, reemplazar todos estos s칤mbolos raros por "/", para conseguir homogeneizar los valores de esta columna. Para ellos deberemos crear una funci칩n.


```python
df_tickets["categor칤a"].unique()
```

```python
patron = "Acces.*"
patron_dos = "Acceso/Login"
```

```python
def cambiar_categoria (patron1, patron2, columna_a_cambiar):
    try:
        cambiado = re.sub(patron1, patron2, columna_a_cambiar)
        return cambiado
    except:
        return np.nan
```

```python
df_tickets["categoria"] = df_tickets.apply(lambda df: cambiar_categoria (patron, patron_dos, df["categor칤a"]), axis = 1)
df_tickets.head()
```
